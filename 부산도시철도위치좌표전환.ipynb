```python
import pandas as pd
df = pd.read_excel('전체_도시철도역사정보_20250417.xlsx', index_col='역번호')
df = pd.DataFrame(df, columns=['역사명', '노선명', '환승역구분', '환승노선명', '역사도로명주소'])
df.index = df.index.astype(str).str.zfill(4)
df = df.query("노선명 in ['부산 도시철도 1호선', '부산 도시철도 2호선', '부산 도시철도 3호선', '부산 경량도시철도 4호선', '부산김해경전철', '동해선']").sort_index()
pd.set_option('display.max_rows', None)
df = df.reset_index()
df # 부산원동역이 안락역, 재송역 사이가 아닌 일광역, 좌천역 사이로 나옴. 실제 순서는 안락역, 재송역 사이가 맞음.
```

```python
import pandas as pd
import requests
import time
import random

apikey = '' # API 키를 여기에 입력
df['위도'] = pd.NA
df['경도'] = pd.NA

for idx, addr in enumerate(df['역사도로명주소']):
    apiurl = 'https://api.vworld.kr/req/address?'
    params = {
        "service": "address",
        "request": "getcoord",
        "crs": "epsg:4326",
        "address": addr,
        "format": "json",
        "type": "road",
        "key": apikey,
    }
    response = requests.get(apiurl, params=params, timeout=2)
    time.sleep(random.uniform(1, 2))  # 1~2초 대기
    if response.status_code == 200:
        response_json = response.json()
        if 'response' in response_json and 'result' in response_json['response'] and 'point' in response_json['response']['result']:
            point = response_json['response']['result']['point']
            df.at[idx, '위도'] = point['y']
            df.at[idx, '경도'] = point['x']
            print(f"인덱스 {idx}, 주소 '{addr}' 출력 성공: 위도 {point['y']}, 경도 {point['x']}")
        else:
            print(f"인덱스 {idx}, 주소 '{addr}' 출력 실패. 재시도")
            # 재시도 (type="parcel")
            params['type'] = 'parcel'
            response = requests.get(apiurl, params=params, timeout=2)
            time.sleep(random.uniform(1, 2))  # 1~2초 대기
            if response.status_code == 200:
                response_json = response.json()
                if 'response' in response_json and 'result' in response_json['response'] and 'point' in response_json['response']['result']:
                    point = response_json['response']['result']['point']
                    df.at[idx, '위도'] = point['y']
                    df.at[idx, '경도'] = point['x']
                    print(f"인덱스 {idx}, 주소 '{addr}' 재시도 성공: 위도 {point['y']}, 경도 {point['x']}")
                    # 출력 후 '재시도도' 오타를 '재시도'로 수정
                else:
                    print(f"인덱스 {idx}, 주소 '{addr}' 출력 실패")
            else:
                print(f"인덱스 {idx}, 주소 '{addr}' 재시도 요청 실패: 상태 코드 {response.status_code}")
    else:
        print(f"인덱스 {idx}, 주소 '{addr}' 요청 실패: 상태 코드 {response.status_code}")

# 결과 확인
print(df[['역번호', '역사명', '노선명', '역사도로명주소' '위도', '경도']].head())

# CSV로 저장 (선택)
df.to_csv('부산광역시_도시철도위치_좌표.csv', index=False, encoding='utf-8')
```

```python
import pandas as pd
import requests
import time
import random

apikey = '' # API 키를 여기에 입력
# df['위도'] = pd.NA
# df['경도'] = pd.NA

for idx, addr in enumerate(df['역사도로명주소']):
    apiurl = 'https://api.vworld.kr/req/address?'
    params = {
        "service": "address",
        "request": "getcoord",
        "crs": "epsg:4326",
        "address": addr,
        "format": "json",
        "type": "parcel",  # 'road' 대신 'parcel'로 변경
        "key": apikey,
    }
    response = requests.get(apiurl, params=params, timeout=2)
    time.sleep(random.uniform(1, 2))  # 1~2초 대기
    if response.status_code == 200:
        response_json = response.json()
        if 'response' in response_json and 'result' in response_json['response'] and 'point' in response_json['response']['result']:
            point = response_json['response']['result']['point']
            df.at[idx, '위도'] = point['y']
            df.at[idx, '경도'] = point['x']
            print(f"인덱스 {idx}, 주소 '{addr}' 출력 성공: 위도 {point['y']}, 경도 {point['x']}")
        else:
            print(f"인덱스 {idx}, 주소 '{addr}' 출력 실패. 재시도")
            # 재시도 (type="parcel")
            params['type'] = 'road'  # 'parcel' 대신 'road'로 변경
            response = requests.get(apiurl, params=params, timeout=2)
            time.sleep(random.uniform(1, 2))  # 1~2초 대기
            if response.status_code == 200:
                response_json = response.json()
                if 'response' in response_json and 'result' in response_json['response'] and 'point' in response_json['response']['result']:
                    point = response_json['response']['result']['point']
                    df.at[idx, '위도'] = point['y']
                    df.at[idx, '경도'] = point['x']
                    print(f"인덱스 {idx}, 주소 '{addr}' 재시도 성공: 위도 {point['y']}, 경도 {point['x']}")
                    # 출력 후 '재시도도' 오타를 '재시도'로 수정
                else:
                    print(f"인덱스 {idx}, 주소 '{addr}' 출력 실패")
            else:
                print(f"인덱스 {idx}, 주소 '{addr}' 재시도 요청 실패: 상태 코드 {response.status_code}")
    else:
        print(f"인덱스 {idx}, 주소 '{addr}' 요청 실패: 상태 코드 {response.status_code}")

# CSV로 저장 (선택)
df.to_csv('부산광역시_도시철도위치_좌표_수정2.csv', index=False, encoding='utf-8')
```
